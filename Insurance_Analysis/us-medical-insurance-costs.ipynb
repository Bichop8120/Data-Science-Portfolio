# U.S. Medical Insurance Costs

```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


insurance_df = pd.read_csv('insurance.csv')
```

```python
class Person:
    def __init__(self,row):
        smoker_dict = {'yes':1, 'no':0}
        sex_dict = {'female':0, 'male':1}
        self.age = int(row[0])
        self.sex = sex_dict[row[1]]
        self.bmi = int(row[2])
        self.children = int(row[3])
        self.smoker = smoker_dict[row[4]]
        self.region = row[5]

    def __str__(self):
        return f'Person is {self.age} years old and is a {self.sex}'
    
    def estimated_insurance_cost(self):
        return float((250*self.age) - (128*self.sex)+370*self.bmi+(425*self.children)+(2400*self.smoker)-12500)

    def estimated_insurance_cost_diff(self, new_person):
        return self.estimated_insurance_cost() - new_person.estimated_insurance_cost()
```

```python
def average_age_with_children(person_dict):
    total_age = 0
    count = 0
    for index,person in person_dict.items():
        if person.children > 0:
            total_age += person.age
            count += 1
    return total_age/count

def average_insurance_cost(person_dict):
    total_cost = 0
    count = 0
    for index,person in person_dict.items():
        total_cost += person.estimated_insurance_cost()
        count += 1
    return total_cost/count

def average_insurance_cost_smokers(person_dict):
    total_cost = 0
    count = 0
    for index,person in person_dict.items():
        if person.smoker == 1:
            total_cost += person.estimated_insurance_cost()
            count += 1
    return total_cost/count

def average_insurance_cost_non_smokers(person_dict):
    total_cost = 0
    count = 0
    for index,person in person_dict.items():
        if person.smoker == 0:
            total_cost += person.estimated_insurance_cost()
            count += 1
    return total_cost/count
```

```python
person_dict = {}

for index, row in insurance_df.iterrows():
    person = Person(row)
    person_dict[index] = person
    
children_avg = average_age_with_children(person_dict)
insurance_avg = average_insurance_cost(person_dict)
insurance_smoker_avg = average_insurance_cost_smokers(person_dict)
insurance_non_smoker_avg = average_insurance_cost_non_smokers(person_dict)
  
print(f'The average age of people with children in our dataset is: {int(children_avg)}')
print(f'The average insurance cost of people in our data set is: ${round(insurance_avg,2)}')
print(f'The average insurance cost of smokers in our data set is: ${round(insurance_smoker_avg,2)}')
print(f'The average insurance cost of non-smokers in our data set is: ${round(insurance_non_smoker_avg,2)}')
```

```python

ax = plt.figure(figsize=(5,3))
plt.bar(['Smokers','Non-Smokers'],[insurance_smoker_avg,insurance_non_smoker_avg])
plt.title('Average Insurance Cost by Smoking Status')
ax.savefig('insurance_cost_smokers_vs_non_smokers.png')

plt.show()
plt.clf()

people_per_region = dict(insurance_df['region'].value_counts())

ax = plt.figure(figsize=(5,3))
plt.bar(people_per_region.keys(),people_per_region.values())
plt.title('Number of People in Each Region')
ax.savefig('age_vs_insurance_cost.png')


plt.show()
plt.clf()

ax = plt.figure(figsize=(5,3))
plt.hist(insurance_df['charges'], bins=25)
plt.title('Distribution of Insurance Charges')
ax.savefig('insurance_charges_distribution.png')

plt.show()
plt.clf()



```
